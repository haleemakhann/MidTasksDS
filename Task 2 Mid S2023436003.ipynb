### 1. What's the probability of a customer having 2 kids

```python
filtered_data = data[data['n_kids'] < 2]
probability = len(filtered_data[filtered_data['n_vg'] > 10]) / len(filtered_data)
print("Probability of a customer owning more than 10 video games given less than 2 kids:", probability)
```

###  4. Estimate the parameters (mean and std deviation) of this normal distribution computationally.  \
### (Bonus: plot the estimated normal distribution on top of the distribution of `n_hours_playing` after removing values < 50)

```python
import numpy as np
fd = data[data['n_hours_playing'] >= 50]
plt.hist(fd['n_hours_playing'], bins=10)
mean = np.mean(fd['n_hours_playing'])
std_dev = np.std(fd['n_hours_playing'])
print("Mean:", mean)
print("Standard Deviation:", std_dev)
x_values = np.linspace(filtered_data['n_hours_playing'].min(), filtered_data['n_hours_playing'].max(), 100)
y_values = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-(x_values - mean)**2 / (2 * std_dev**2))

#Histogram
plt.hist(filtered_data['n_hours_playing'], bins=10, density=True, alpha=0.5, label='Data')

# Plot the estimated normal distribution
plt.plot(x_values, y_values, color='red', label='Normal Distribution')

# Customize the plot
plt.xlabel('Number of Hours Played')
plt.ylabel('Density')
plt.title('Distribution of Hours Played (Values >= 50)')
plt.legend()

# Show the plot
plt.show()
```

```python
df_person_data.dropna(subset=['age'], inplace=True)
X = df_person_data.drop(['person_id','n_hours_playing'], axis=1)  # Input features
y = df_person_data['n_hours_playing']
from sklearn import preprocessing
x1 = X.values
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x1)
df_scaled = pd.DataFrame(x_scaled)
df_scaled.head()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

from sklearn.metrics import r2_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, matthews_corrcoef

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df_scaled.values, y, test_size=0.2, random_state=42)

# Create a linear regression model
linear_reg = LinearRegression()

# Fit the model to the training data
linear_reg.fit(X_train, y_train)

# Predict on the test set
y_pred = linear_reg.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print('Mean Squared Error:', mse)
mae = mean_absolute_error(y_test, y_pred)
print('Mean Absolute Error:', mae)
r2 = r2_score(y_test, y_pred)
print('R2 Score:', r2)
```
